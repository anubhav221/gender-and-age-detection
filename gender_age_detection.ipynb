```python
import cv2
import numpy as np
%matplotlib inline
%config Completer.use_jedi = False
```

```python

age_list = ['(0, 12)', '(4, 6)', '(8, 12)', '(15, 20)', '(25, 32)', '(38, 43)', '(48, 53)', '(60, 100)']
gender_list = ['Male', 'Female'] 
```

```python
age_net = cv2.dnn.readNetFromCaffe('deploy_age.prototxt', 'age_net.caffemodel')
gender_net = cv2.dnn.readNetFromCaffe('deploy_gender.prototxt', 'gender_net.caffemodel')
```

```python
image = cv2.imread('IMG_9923.JPG')
cv2.imshow('grp_pic',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
small = cv2.resize(image, dsize=None, fx=0.5, fy=0.5)
cv2.imshow('grp_pic',small)  
cv2.waitKey(0) 
cv2.destroyAllWindows()  
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
gray = cv2.cvtColor(small, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 6) 
```

```python
faces
```

```python
if(len(faces)>0):
    print("Found {} faces".format(str(len(faces))))
```

```python
for (x, y, w, h )in faces:
    cv2.rectangle(small, (x, y), (x+w, y+h), (255, 255, 0), 2)

#Get Face 
    face_img = small[y:y+h, h:h+w].copy()
    MODEL_MEAN_VALUES = (np.mean(face_img[:,:,0]), np.mean(face_img[:,:,1]), np.mean(face_img[:,:,2])
    blob = cv2.dnn.blobFromImage(face_img, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
    print(blob.shape)
#Predict Gender
    gender_net.setInput(blob)
    gender_preds = gender_net.forward()
    print('gender_preds - ')
    print(gender_preds)
    gender = gender_list[gender_preds[0].argmax()]
    print("Gender : " + gender)

#Predict Age
    age_net.setInput(blob)
    age_preds = age_net.forward()
    print('age_preds - ')
    print(age_preds)
    age = age_list[age_preds[0].argmax()]
    print("Age Range: " + age) 
    
    cv2.putText(small, str(gender)+', '+str(age), (x, y),cv2.FONT_HERSHEY_COMPLEX 
                    ,1, (255, 255, 255), 2, cv2.LINE_AA) 
```

```python
def detect_face(frame): 
    age_list = ['(0, 3)', '(3, 9)', '(9, 18)', '(18, 30)', '(30, 45)', '(45, 62)', '(62, 80)', '(80, 100)']
    gender_list = ['Male', 'Female'] 
    
    age_net = cv2.dnn.readNetFromCaffe('deploy_age.prototxt', 'age_net.caffemodel')
    gender_net = cv2.dnn.readNetFromCaffe('deploy_gender.prototxt', 'gender_net.caffemodel')
        
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 2) 
    print("no of faces : "+str(len(faces)))
    
    if len(faces)>0 :
        for (x, y, w, h )in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 255, 0), 3)
            #Get Face 
            face_img = frame[y:y+h, h:h+w].copy()  
            MODEL_MEAN_VALUES = (np.mean(face_img[:,:,0]), np.mean(face_img[:,:,1]), np.mean(face_img[:,:,2]))
            blob = cv2.dnn.blobFromImage(face_img, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
                #print(blob.shape)
            #Predict Gender
            gender_net.setInput(blob)
            gender_preds = gender_net.forward()
                #print('gender_preds - ')
                #print(gender_preds)
            gender = gender_list[gender_preds[0].argmax()]
            print("Gender : " + gender) 

            #Predict Age
            age_net.setInput(blob)
            age_preds = age_net.forward() 
            print('age_preds - ')
            print(age_preds)
            age = age_list[age_preds[0].argmax()]
            print("Age Range: " + age) 

            cv2.putText(frame, str(gender)+', '+str(age), (x, y),cv2.FONT_HERSHEY_COMPLEX 
                                ,1, (255, 255, 255), 2, cv2.LINE_AA)
            return frame
        else:
            return frame
    
        
```

```python
cap = cv2.VideoCapture(0)

cap.set(3, 480) #set width of the frame
cap.set(4, 640) #set height of the frame
```

```python
while True:
    ret, image = cap.read()
    if ret == True: 
        frame = detect_face(image)
        cv2.imshow('image', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'): 
            cv2.destroyAllWindows()
            break 
```

```python
def frame(image,resize = False):
    image = cv2.imread('IMG_9923.JPG')
    if resize == True :
        small = cv2.resize(image, dsize=None, fx=0.5, fy=0.5)
    else:
        small = image
    return small
```
